```python
# Find Missing values in Data or finding NA values

dataframe.isna()
dataframe.isna().sum() # Values for each columns says that this number of missing values are there.
```

```python
# Taking a dataframe where NA(null) values are removed
na_data_removed=dataframe.dropna()
na_data_removed.shape
```

```python
# Here we can see the difference that comes in the available data if we remove all those rows thta have null values or NA values
dataframe.shape 
```

```python
# Taking a dataframe where NA(null) values are removed based on Columns if they are null
na_data_removed_col=dataframe.dropna(axis=1)
na_data_removed_col.shape
```

```python
# We can't drop more than 8%-10% data
# We need to fill NA by 0s 

na_data_fill=dataframe.fillna(0)
na_data_fill.shape
```

```python
# Values for each columns says that this number of missing values are there when we fill it with fillna()
# Now we will get no missing values
na_data_fill.isna().sum() 
```

```python
na_data_fill.head()
```

```python
# We can't drop more than 8%-10% data
# We need to fill NA by filling values using : pad/ffill: propagate last valid observation forward
# backfill/bfill: Use next valid observation to fill gap

na_data_backfill=dataframe.fillna(method='bfill')
na_data_backfill.shape
```

```python
na_data_backfill.head()
```

```python
# If sometimes ffill that fill values of previous valid result to next NA. In case where it is the first record in the 
# column records then we need to fill by adding fillna(method = 'bfill')
# We need to fill NA by filling values using : pad/ffill: propagate last valid observation forward
# backfill/bfill: Use next valid observation to fill gap

na_data_ffill=dataframe.fillna(method='ffill')
na_data_ffill.isna().sum()
```

```python
# If sometimes ffill that fill values of previous valid result to next NA. In case where it is the first record in the 
# column records then we need to fill by adding fillna(method = 'bfill')
# We need to fill NA by filling values using : pad/ffill: propagate last valid observation forward
# backfill/bfill: Use next valid observation to fill gap

na_data_ffill=dataframe.fillna(method='ffill').fillna(method='bfill')
na_data_ffill.isna().sum()
na_data_ffill.head()
```

```python
# Below will remove the duplicates when show_id, cast and director is same in different record and identify then as Duplicates
drop_duplicates= dataframe.drop_duplicates(subset=['cast','director'])
print(dataframe.shape,drop_duplicates.shape)
```
