```python
# Data Cleaning in 
import pandas as pd
data = pd.read_csv("netflix_titles.csv")

data.shape
```

```python
data.info()
```

```python
# Fill NUll values whereever in dataset
na_data_ffill=data.fillna(method='ffill').fillna(method='bfill')
```

```python
# Here we are using REGULAR EXPRESSION LIBRARY of Python for finding pattern/ Splitting Record/ Replacing values
# Importing Regular Expression Library
import re

# Finding in a string a string using regular expression : re.findall("String", a)
a="Hello Welcome to the class. This is awesome class"
re.findall("class",a)

#Splitting line using regular expression using a special character (!, &, #)
b="This world is full of Wonders # Taj Mahal in India is one of them "
re.split("#",b)
```
